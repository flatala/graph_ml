```python
from pathlib import Path
import sys

ROOT = Path.cwd().parent
sys.path.append(str(ROOT))

import os
import pandas as pd
import numpy as np
```

```python
DATA_DIR = "../../elliptic_dataset"
WALLETS = "wallets_features_until_t.csv"

nodes = pd.read_csv(os.path.join(DATA_DIR, WALLETS))

print(f"Nodes: {len(nodes):,}")
```

```python
labels = pd.read_csv(os.path.join(DATA_DIR, "wallets_classes.csv"))
print(len(labels))
nodes_with_classes = nodes.merge(labels, on='address', how='left')
print(nodes_with_classes.shape[0])
```

```python
time_steps = sorted(nodes_with_classes["Time step"].unique())
print(time_steps)
```

```python
for time_step in time_steps:
    print(f"\ntime step: {time_step}")
    nodes_at_time_step = nodes_with_classes[nodes_with_classes["Time step"] == time_step]
    num_total = nodes_at_time_step.shape[0]
    num_illicit = nodes_at_time_step[nodes_at_time_step["class"] == 1].shape[0]

    print(f"total: {num_total} | illicit: {num_illicit} | ratio: {num_illicit / num_total}")
    
```
